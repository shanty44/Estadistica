Mostrar una dataframe con el nombre del barrio, el precio y el número de entradas.

Ayuda: Podeis crear un nuevo dataframe con las columnas "Barrio" y "Freq" que contenga el número de entradas en cada barrio y hacer un merge con el dataframe del punto anterior.

```r
library(dplyr)
library(tidyr)

barrio_frec <- data.frame(summary(airbnb_entire$Barrio))
barrio_frec$Barrio <- row.names(barrio_frec)
row.names(barrio_frec) <- NULL
colnames(barrio_frec) <- c('Frecuencia','Barrio')
barrio_frec_precio <- merge(airbnb_entire_ag,barrio_frec)
barrio_frec_precio[order(barrio_frec_precio$Frecuencia, decreasing=T),]
```

```r
calculo = cor(barrio_sol[,c("NumBanyos", "NumDormitorios", "MaxOcupantes", "MetrosCuadrados")], use = "pairwise.complete.obs")
calculo
```

Se observa que la correlación entre el número de dormitorios y los metros cuadrados es sorprendentemente baja.
¿Son de fiar esos números?

Mediante un histograma o curvas de densidad podemos descartar números que notienen sentido en el dataframe barrio_sol,
para tener una matriz de correlación que tenga mayor sentido.


```r
options(repr.plot.height=3,repr.plot.width=3)
ggplot(data = barrio_sol, aes(x=MetrosCuadrados)) + 
    geom_histogram(bins=15) + 
    xlab("Metros Cuadrados")
ggplot(data = barrio_sol, aes(x=NumDormitorios)) + 
    geom_histogram(bins=15) + 
    xlab("Cantidad dormitorios")
```

### Hay que descartar los departamentos que tienen menos de 20 mts2 y los que tienen 0 dormitorios

Una vez que hayamos filtrado los datos correspondientes calcular el valor o la combinación de valores que mejor nos permite obtener el precio de un inmueble.

```r
calculo <- lm(data=barrio_sol %>%
              filter(MetrosCuadrados > 20 & NumDormitorios >= 1), formula = Precio ~ MetrosCuadrados+NumDormitorios+NumBanyos+MaxOcupantes)
summary(calculo)
confint(calculo)
```
